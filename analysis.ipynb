## Data Review and information 

```python
# Create an empty dictionary to store unique values for each column
unique_values = {}

# Loop through the columns and find unique values
for column in feb_2020.columns:
    unique_values[column] = feb_2020[column].unique()

# Display unique values for each column
for column, values in unique_values.items():
    print(f'Column: {column}')
    print(values)
    print('\n')
```

```python
# Create an empty dictionary to store unique values for each column
unique_values = {}

# Loop through the columns and find unique values
for column in march_2020.columns:
    unique_values[column] = march_2020[column].unique()

# Display unique values for each column
for column, values in unique_values.items():
    print(f'Column: {column}')
    print(values)
    print('\n')
```

```python
# Remove all birth year from 1920- assuming all data points where age >100 is a wrong input or possible type. 
feb_2020_df = feb_2020[(feb_2020['birth year'] >= 1920) & (feb_2020['birth year'] <=2020)]

# Change gender to correct string 
feb_2020_df['gender'] = feb_2020_df['gender'].replace({1: 'Male', 2: 'Female', 0: 'Unknown'}) 

feb_2020_df.to_csv('Output/clean_feb_2020.csv', index=False) 
```

```python
# Remove all birth year from 1920- assuming all data points where age >100 is a wrong input or possible type. 
march_2020_df = march_2020[(march_2020['birth year'] >= 1920) & (march_2020['birth year'] <=2020)]

# Change gender to correct string
march_2020_df['gender'] = march_2020_df['gender'].replace({1: 'Male', 2: 'Female', 0: 'Unknown'}) 

march_2020_df.to_csv('Output/clean_march_2020.csv', index=False) 
```

```python
# Merge both data to have a comparison between months 
merged_df = pd.merge(feb_2020_df, march_2020_df, how='outer') 

# Create csv of merged df 
merged_df.to_csv('Output/merged_feb_march_2020.csv', index=False)  
```
